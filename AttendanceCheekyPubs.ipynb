```python
from datetime import date, datetime, time, timedelta
import pandas as pd
data = pd.read_csv('data.txt', sep="\t+", engine='python')
data.columns = ["No","Mchn","EnNo","Name","Mode","IOMd","DateTime"]

data['DateTime'] = pd.to_datetime(data['DateTime'])
data = data.iloc[::-1] 
data['Dates'] = pd.to_datetime(data['DateTime']).dt.date
data['Time'] = pd.to_datetime(data['DateTime']).dt.time

exampleDt = timedelta()
dataEmployee = pd.DataFrame()
employees = dict()

for index, row in data.iterrows():
  if row.Name in employees:
    employees[row.Name].append({"date" : row.Dates, "time": row.Time, "datetime": row.DateTime})
  else:
    employees[row.Name] = [{"date" : row.Dates, "time": row.Time, "datetime": row.DateTime}]



for item in employees:
  #print(employees[item])

  empDisplay = dict()
  empColumns = dict()

  for dates in employees[item]:
    #empDisplay.append({dates.get('date'):dates.get('time')})
    
    if dates.get('date') in empDisplay:
      empDisplay[dates.get('date')].append(dates.get('time'))

      if empColumns.get('count') == 0:
        dtime = empColumns.get('dateOld') - dates.get('datetime')
        empColumns[dates.get('date')] = empColumns[dates.get('date')] + dtime
        empColumns['count'] = 1
      else: 
        empColumns['dateOld'] = dates.get('datetime')
        empColumns['count'] = 0
    else:
      empDisplay[dates.get('date')] = [dates.get('date'), dates.get('time')]
      empColumns[dates.get('date')] = exampleDt
      empColumns['count'] = 0
      empColumns['dateOld'] = dates.get('datetime')

  del empColumns["dateOld"]
  del empColumns["count"]
  e = pd.DataFrame(list(empDisplay.values()))
  e = e.set_index(e.columns[0])
  e['Hours'] = list(empColumns.values())
  e = e.iloc[::-1, ::-1]
  e.index.name = None
  e.columns.name = item.upper()
  display(e)
  total = round(e['Hours'].sum() / pd.Timedelta(hours=1), 2)
  print()
  print(item.upper().strip(),"- Total Hours:", total)
  print()
  print('----------------------------------')
  print()
  #display(e.iloc[:, ::-1])

print()
print('All data: ')
print()
data = pd.read_csv('data.txt', sep="\t+", engine='python')
data.columns = ["No","Mchn","EnNo","Name","Mode","IOMd","DateTime"]
#data = data.iloc[::-1]
#data['DateTime'] = pd.to_datetime(data['DateTime'])
data['Dates'] = pd.to_datetime(data['DateTime']).dt.date
data['Time'] = pd.to_datetime(data['DateTime']).dt.time

display(data)
```
